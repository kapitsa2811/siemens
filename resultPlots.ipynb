```python
filePath=".//data//"
modelPath=".//model//"
plotPath=".//plots//"
fileName="data_sample.csv"
resultPath=".//results//"

def readData(path):
    df1 = pd.read_csv(path)
    df1=df1.dropna()
    print("\n\t shape2:", df1.shape)
    return df1
path=resultPath+"outlierVoting_test.csv"
df=readData(path)
print("\n\t shape: df1:",df.shape)
```
